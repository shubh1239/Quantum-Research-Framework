```python
import numpy as np 
```

```python
X = np.array([[0,1],[1,0]])
Y = np.array([[0,-1j],[1j,0]])
X_tensor_Y = np.kron(X,Y) #kronecker product or tensor product multiplies each element of the first matrix by the second matrix
print(X_tensor_Y)
```

```python
#operator function exponentiation
from scipy.linalg import expm
Z= np.array([[1,0],[0,-1]])
Z_exp = expm(Z) #exponential of a matrix
print(Z_exp)
```

```python
#trace of a matrix is the sum of its diagonal elements
trace_Z = np.trace(Z)
print(trace_Z)
```

```python
commutator_XY = np.dot(X,Y) - np.dot(Y,X) #commutator of two matrices is defined as [A,B] = AB - BA
print(commutator_XY)
```
