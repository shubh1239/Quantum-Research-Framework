```python
import numpy as np
```

```python
#Eigen values of hamiltonian matrix are purely real numbers, which represent the energy levels of the system. The corresponding eigenvectors represent the energy eigenstates of the system. In this case, we have two energy levels: 1 and -1, with their corresponding eigenvectors representing the states of the qubit.
```

```python
# Define the Hamiltonian for a simple two-level system (qubit)
#hamiltonian matrix represents the energy levels of the system
#In this case, we have a two-level system with energy levels 1 and -1, which can be represented by the Pauli Z matrix
H = np.array([[1, 0], [0, -1]]) #Pauli Z matrix
# The eigenvalues of the Hamiltonian represent the energy levels of the system
eigenvalues, eigenvectors = np.linalg.eig(H)
print("Eigenvalues (Energy levels):", eigenvalues)
print("Eigenvectors (Energy eigenstates):\n", eigenvectors)
print(f"Is the Hamiltonian matrix Hermitian? {np.allclose(H, H.conj().T)}")
print(f"Is the Hamiltonian matrix unitary? {np.allclose(H @ H.conj().T, np.eye(H.shape[0]))}")
print("Ground state energy:", np.min(eigenvalues))

```
