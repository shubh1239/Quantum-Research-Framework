```python
import numpy as np
```

```python
A = np.array([[1,2],[3,4]])

```

```python
#lets perform Singular Value Decomposition (SVD) on the matrix A
U , D , Vt = np.linalg.svd(A) #U is the left singular vectors, D is the singular values, and Vt is the right
#np.linalg.svd returns the singular values in descending order, so we can use them to reconstruct the original matrix A
#we can also use the singular values to perform dimensionality reduction by keeping only the top k singular values and their corresponding singular vectors
```

```python
D = np.diag(D) #convert the singular values into a diagonal matrix
A_reconstructed = np.dot(U, np.dot(D, Vt)) #reconstruct the original matrix A using the singular values and singular vectors
print("Original Matrix A:")
print(A)
print("Reconstructed Matrix A:")
print(A_reconstructed)
print("Left Singular Vectors (U):")
print(U)
print("Singular Values (D):")
print(D)
print("Right Singular Vectors (Vt):")
print(Vt)
```

```python
print(f"is the original matrix A equal to the reconstructed matrix A? {np.allclose(A, A_reconstructed)}")
```
