```python
import numpy as np
```

```python
#distinguishing quantum states
# alice sends one of two states to bob, either |0> or |1>
# bob wants to determine which state he received
# if bob measures in the computational basis, he will always get the correct answer
#|0> = [1, 0] , |1> = [0, 1] are orthonal states, so they can be perfectly distinguished
state_0 = np.array([1, 0])
state_1 = np.array([0, 1])
# bob measures in the computational basis
measurement_0 = np.array([1, 0])
measurement_1 = np.array([0, 1])
# probability of getting outcome 0 when measuring state_0
prob_0_given_0 = np.abs(np.dot(measurement_0, state_0))**2
# probability of getting outcome 1 when measuring state_0
prob_1_given_0 = np.abs(np.dot(measurement_1, state_0))**2
# probability of getting outcome 0 when measuring state_1
prob_0_given_1 = np.abs(np.dot(measurement_0, state_1))**2
# probability of getting outcome 1 when measuring state_1
prob_1_given_1 = np.abs(np.dot(measurement_1, state_1))**2
print("Probability of getting outcome 0 when measuring state_0:", prob_0_given_0)
print("Probability of getting outcome 1 when measuring state_0:", prob_1_given_0)
print("Probability of getting outcome 0 when measuring state_1:", prob_0_given_1)
print("Probability of getting outcome 1 when measuring state_1:", prob_1_given_1)
```
