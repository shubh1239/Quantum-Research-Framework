```python
import numpy as np 
```

```python
# 1. Define the Initial State sent by Alice: |+> state
psi_alice = (1 / np.sqrt(2)) * np.array([1, 1])
print("Alice sends State |+> :", psi_alice)
```

```python
# 2. Eve intercepts using the WRONG basis (Computational Basis Matrix M0 = |0><0|)
M0 = np.array([[1, 0],
               [0, 0]])
```

```python
# Probability of Eve collapsing the state to |0>
prob_eve = psi_alice.T @ M0 @ psi_alice
print(f"Probability of Eve collapsing the state: {prob_eve * 100:.1f}%")
```

```python

# State after Eve's measurement collapse: System becomes perfectly |0>
psi_collapsed = np.array([1, 0])
print("State after Eve's intercept (Collapsed to |0>):", psi_collapsed)
```

```python
# 3. Bob measures the collapsed state in Alice's original basis (Projector to |->)
# Filter matrix for |-> state to check for errors
E_minus = 0.5 * np.array([[1, -1],
                          [-1, 1]])
```

```python
# Bob checks if he detects an error (ideally for |+>, probability of |-> should be 0%)
bob_error_prob = psi_collapsed.T @ E_minus @ psi_collapsed
```

```python
print(f"Probability of Bob catching an Error: {bob_error_prob * 100:.1f}%")
print("Conclusion: Eve's wrong guess caused a collapse, injecting 50% error for Bob. CAUGHT!")
```
