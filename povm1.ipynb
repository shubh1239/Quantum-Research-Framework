```python
import numpy as np

# 1. Define POVM elements with explicit 2/3 scaling factor applied correctly
m1 = (2/3) * np.array([[1, 0], 
                       [0, 0]])

m2 = (1/3) * np.array([[0.5, np.sqrt(3)/2], 
                       [np.sqrt(3)/2, 1.5]])

m3 = (1/3) * np.array([[0.5, -np.sqrt(3)/2], 
                       [-np.sqrt(3)/2, 1.5]])

# 2. Strict Identity Audit
assert np.allclose(m1 + m2 + m3, np.eye(2)), "POVM elements do not sum to the identity!"
print("✅ SUCCESS: POVM completeness condition satisfied! Sum is Identity Matrix.\n")

# 3. Define State |+><+| density matrix
psi = (1/2) * np.array([[1, 1], 
                        [1, 1]])

# 4. Compute Tr(E_i * rho) via Matrix Overlaps
p1 = np.trace(m1 @ psi)
p2 = np.trace(m2 @ psi)
p3 = np.trace(m3 @ psi)

# 5. Print Results
print("--- 📊 POVM MEASUREMENT DISTRIBUTIONS ---")
print(f"Probability of outcome 1 (|0> component)   : {p1:.4f}")
print(f"Probability of outcome 2 (120° component) : {p2:.4f}")
print(f"Probability of outcome 3 (240° component) : {p3:.4f}")
print(f"Total Combined Probability                 : {p1 + p2 + p3:.4f}")
```
