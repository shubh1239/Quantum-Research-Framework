```python
import numpy as np 
```

```python
#define pauli matrices
X = np.array([[0,1],[1,0]])
Z = np.array([[1,0],[0,-1]])
Y = np.array([[0,-1j],[1j,0]])
```

```python
#commutator of two matrices is defined as [A,B] = AB - BA
XZ = np.dot(X,Z) 
ZX = np.dot(Z,X)
commutator_XZ = XZ - ZX
RHS = -2 * 1j *Y
```

```python
print("Commutator [X,Z]:")
print(commutator_XZ)
print("Right-hand side 2iY:")
print(RHS)
print("Are they equal?", np.allclose(commutator_XZ, RHS)) #np.allclose checks if two arrays are element-wise equal within a tolerance. This is important for comparing floating-point numbers, which may not be exactly equal due to precision issues.
```
