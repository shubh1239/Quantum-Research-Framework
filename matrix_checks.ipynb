```python
import numpy as np
```

```python
Z = np.array([[1,0],[0,-1]])
Z_ = np.conj(Z).T
```

```python
#check if hermitian 
hermitian = np.allclose(Z,Z_) #allclose checks if two arrays are element-wise equal within a tolerance.
print(f"is pauli-z hermitian ? {hermitian}")  
```

```python
identity = np.dot(Z,Z_) 
unitary = np.allclose(identity, np.eye(2)) #np.eye(2) creates a 2x2 identity matrix. The np.allclose function checks if the product of Z and its conjugate transpose is close to the identity matrix, which is a condition for unitarity.
print(f"is pauliz unitary {unitary}")
```
